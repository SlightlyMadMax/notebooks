```python
import numpy as np
from scipy.sparse import diags
```

```python
n_y, n_x = 5, 5
H = W = 1
h_x, h_y = W / (n_x - 1), H / (n_y - 1)
x = np.linspace(0, W, n_x)
y = np.linspace(0, H, n_y)
```

```python
inner_n_y, inner_n_x = n_y - 2, n_x - 2
size = inner_n_x * inner_n_y

main_diag = -2 / h_x**2 - 2 / h_y**2
off_diag_x = 1 / h_x**2
off_diag_y = 1 / h_y**2

diagonals = [
    np.full(size, main_diag),  # main
    np.full(size - 1, off_diag_x),  # left/right
    np.full(size - 1, off_diag_x),
    np.full(size - inner_n_x, off_diag_y),  # top/bottom
    np.full(size - inner_n_x, off_diag_y),
]
offsets = [0, -1, 1, -inner_n_x, inner_n_x]

for i in range(1, inner_n_y):
    diagonals[1][i * inner_n_x - 1] = 0.0
    diagonals[2][i * inner_n_x] = 0.0
    
A = diags(diagonals, offsets, shape=(size, size), format="csr")
A.todense()
```
